# Rank Corelation

```python
x=list(map(int,input().split()))
y=list(map(int,input().split()))
ranksx=[]
ranksy=[]
rx=[]
ry=[]
xsort=sorted(x,reverse=True)
ysort=sorted(y,reverse=True)
def ran(sort,ranks):
  k=0
  c=0
  c1=0
  sum1=0
  while(k<len(sort)):
    if(sort.count(sort[k])!=1):
      a=sort.count(sort[k])  
      while(c<a):
        k=k+1
        sum1=sum1+k
        c=c+1
      while(c1<a):
        rank=sum1/a
        ranks.append(rank)
        c1 =c1+1
    else:
      k+=1
      ranks.append(k)
    c=0
    c1=0
    sum1=0
  return ranks
ranksx=ran(xsort,ranksx)
ranksy=ran(ysort,ranksy)
def ran1(ranks,sort,a):
  k=0
  r=[]
  while(k<len(ranks)):
    j=a[k]
    r.append(ranks[sort.index(j)])
    k+=1
  return r
rx=ran1(ranksx,xsort,x)
ry=ran1(ranksy,ysort,y)
print(x)
print(rx)
print(y)
print(ry)
di=[]
disqr=[]
n=len(x)
for i in range(0,n):
    di.append(rx[i]-ry[i])
    disqr.append(di[i]*di[i])
print(disqr)
sdisqr=sum(disqr)
print(sdisqr)
repetedx=[]
for i in x:
    if(x.count(i)>1 and i not in repetedx):
        repetedx.append(i)
countsx=[]
for i in repetedx:
    c=x.count(i)
    countsx.append(c)
sumcfx=0
for i in countsx:
    s=(i*i*i-i)/12
    sumcfx=sumcfx+s
repetedy=[]
for i in y:
    if(y.count(i)>1 and i not in repetedy):
        repetedy.append(i)
countsy=[]
for i in repetedy:
    c=y.count(i)
    countsy.append(c)
sumcfy=0
for i in countsy:
    s=(i*i*i-i)/12
    sumcfy=sumcfy+s
print(repetedx)
print(repetedy)
sdisqr=sdisqr+sumcfx+sumcfy
print(sdisqr)
s=1-((6*sdisqr)/(n*((n**2)-1)))
print("Spearmanns Rank Coefficient:",s)
```
